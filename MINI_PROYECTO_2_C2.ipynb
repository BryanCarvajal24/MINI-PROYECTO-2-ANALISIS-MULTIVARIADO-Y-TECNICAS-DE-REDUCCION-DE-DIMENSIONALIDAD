Rescatar valores

```r
total_localidad_validas <- sum(
  df_caso2$COMUNA == "Rural" &
    !(
      toupper(str_trim(df_caso2$LOCALIDAD)) %in% c("URBANA", "RURAL", "") |
      is.na(df_caso2$LOCALIDAD)
    ),
  na.rm = TRUE
)


vals_localidad <- str_trim(df_caso2$LOCALIDAD[
  df_caso2$COMUNA == "Rural" &
    !(
      toupper(str_trim(df_caso2$LOCALIDAD)) %in% c("URBANA", "RURAL", "") |
      is.na(df_caso2$LOCALIDAD)
    )
])


cat("Total de filas con COMUNA='Rural' y LOCALIDAD válida:", total_localidad_validas, "\n\n")
cat("Valores de LOCALIDAD válidas:\n", paste(vals_localidad, collapse = "\n"), "\n")
```

```r

# Filtro de filas donde COMUNA = "Rural" y LOCALIDAD es válida
mask_valid <- df_caso2$COMUNA == "Rural" &
  !(toupper(str_trim(df_caso2$LOCALIDAD)) %in% c("URBANA", "RURAL", "") |
    is.na(df_caso2$LOCALIDAD))

# Extraer número de la columna LOCALIDAD (manteniendo como texto)
nums_comuna <- str_extract(str_trim(df_caso2$LOCALIDAD[mask_valid]), "\\d+")

# Reemplazar valores en la columna COMUNA
df_caso2$COMUNA[mask_valid] <- nums_comuna


cat("Total de reemplazos realizados:", length(nums_comuna), "\n\n")
cat("Filas actualizadas (columna COMUNA):\n")


print(df_caso2[mask_valid, c("LOCALIDAD", "COMUNA")])
```

Imputación comuna

```r
df_caso2$COMUNA <- as.factor(df_caso2$COMUNA)
```

```r
df_caso2$COMUNA[df_caso2$COMUNA == "Rural"] <- NA
```

```r
sum(is.na(df_caso2$COMUNA))
```

```r
# --- Imputación múltiple para COMUNA ---

# Seleccionamos solo las variables relevantes para predecir COMUNA
vars_imputar <- c("COMUNA", "ESTRA_MODA", "DINÁMICA", "CARACTERIZ",
                  "MODALIDAD", "HORA_DEL_H", "DÍA_SEMAN", "MES_DEL_HE")

# Subconjunto del dataframe
df_imput <- df_caso2[, vars_imputar]

imp <- mice(df_imput, m = 5, method = "polyreg", seed = 123)

df_completo <- complete(imp, 1)

# Reasignar la columna imputada al dataframe original
df_caso2$COMUNA <- df_completo$COMUNA


sum(is.na(df_caso2$COMUNA))
```

```r
g1 <-stripplot(imp, COMUNA ~ .imp,
           pch = 20,
           cex = 1.2,
           col = c("skyblue", "tomato"),
           main = "Distribución de imputaciones para la variable COMUNA")


print(g1)
```

```r
# Crear un dataframe combinado con etiquetas "Original" e "Imputado"
df_compare <- data.frame(
  COMUNA = c(df_imput$COMUNA, df_completo$COMUNA),
  Tipo = rep(c("Original", "Imputado"), each = nrow(df_imput))
)

# Gráfico de barras comparativo
g2 <- ggplot(df_compare, aes(x = COMUNA, fill = Tipo)) +
  geom_bar(position = "dodge") +
  labs(title = "Comparación de frecuencias de COMUNA: original vs imputada",
       x = "Comuna",
       y = "Frecuencia",
       fill = "Tipo de dato") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))


print(g2)
```
